```python
#bai6
import osmnx as ox
import folium

placee='District 10, Ho Chi Minh City, Vietnam'
graph= ox.graph_from_place(placee, network_type='drive')

tk = ox.basic_stats(graph)

print('THỐNG KÊ')
print(f'Các nút: {tk['n']}')
print(f'Các đoạn đường: {tk['m']}')
print(f'Độ dài tb: {round(tk['street_length_avg'],2)} m')

nut,dg=ox.graph_to_gdfs(graph)

center=[nut.geometry.y.mean(),nut.geometry.x.mean()]
m=folium.Map(location=center, zoom_start=14)

for _, row in dg.iterrows():
    coords = [(y, x) for x, y in row.geometry.coords]
    folium.PolyLine(coords,color='blue').add_to(m)

sample_nodes=nut.sample(min(300, len(nut)))

for _, row in sample_nodes.iterrows():
    folium.CircleMarker(location=(row.geometry.y, row.geometry.x),radius=3,color='red',fill=True).add_to(m)
m
```
